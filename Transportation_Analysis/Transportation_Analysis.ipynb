## Step 3: Replacing Column Headers with Abbreviations

This step is to replace the column headers with abbreviations to help shorten the names of the columns. Some of the column names are too long so abbreviating things like Highway to hwy or International to intl could make it much more visually appealing. 

```python
# Define common abbreviations for some of the words used in the headers
abbreviations = {
    'Highway': 'Hwy',
    'Vehicle': 'Veh',
    'Miles': 'Mi',
    'Traveled': 'Trav',
    'Fatalities': 'Fatl',
    'Passenger': 'Pass',
    'Rail': 'Rl',
    'Freight': 'Frgt',
    'Traffic': 'Trfc',
    'International': 'Intl',
    'Domestic': 'Dom',
    'Ridership': 'Rider',
    'Transit': 'Trans',
    'Adjusted': 'Adj',
    'Intermodal': 'Intermod',
    'Reports': 'Rpt',
    'Total': 'Tot'
}

# Method which takes in headers and the abbreviations and then applies the abbreviations to the header. 
def abbreviate_headers(headers, abbreviations):
    abbreviated_headers = []
    for header in headers:
        for word, abbr in abbreviations.items():
            header = header.replace(word, abbr)
        abbreviated_headers.append(header)
    return abbreviated_headers

# Get the current headers
current_headers = df.columns.tolist()

# Abbreviate the headers
abbreviated_headers = abbreviate_headers(current_headers, abbreviations)

# Create a mapping dictionary from current to abbreviated headers
header_mapping = dict(zip(current_headers, abbreviated_headers))

# Apply the mapping to rename columns
df.rename(columns=header_mapping, inplace=True)

# Verify the changes
df.head()
```

## Step 4: Removing non-alphanumeric symbols

This step is to remove the non-alphanumeric symbols from the column headers, in turn shortening the labels for each. 

```python
# This is a function that will help remove the non-alphanumeric symbols from the headers. 
def simp_header(df):
    """
    Generates a simplified column mapping for a DataFrame with capitalized words.
    
    Parameters:
    df (pd.DataFrame): The DataFrame with original headers.
    
    Returns:
    dict: A dictionary with original column names as keys and simplified column names as values.
    """
    mapping = {}
    for col in df.columns:
        simplified_col = col.replace(' - ', ' ').replace(' ', '_').replace('/', '_').replace(',', '').replace('__', '_').replace('-', ' ').title().replace('_', ' ')
        mapping[col] = simplified_col
    return mapping

# Generate the mapping
simp_header_mapping = simp_header(df)

# Function to apply the mapping of simplified headers. 
def apply_simplified_headers(df, mapping):
    df.rename(columns=mapping, inplace=True)
    return df

# Apply the function to simplify headers
df = apply_simplified_headers(df, simp_header_mapping)

# Verify the changes
df.head()
```

## Step 5: Removing Completely Empty Rows/Columns

This step removes the Date column and checks all remaining rows to see if any are completely empty. If they are, we proceed to remove them from the dataset in turn shortening our dataset. 

```python
# Create a subset of the data excluding the 'Date' column
subset_df = df.drop(columns=['Date'])

# Identify rows where all other columns are empty
empty_rows = subset_df.isnull().all(axis=1)

# Save the indices of completely empty rows
empty_row_indices = df[empty_rows].index

# Drop these empty rows from the original data
df = df.drop(index=empty_row_indices)

# Verify changes (dataset should be smaller)
df.shape
```

```python
# Identify columns where all values are missing
empty_columns = subset_df.isnull().all(axis=0)

# Print the names of completely empty columns
empty_columns_list = subset_df.columns[empty_columns]
print("Completely empty columns (excluding 'Date' column):", empty_columns_list)

# Drop empty columns, if any
df = df.drop(columns=empty_columns_list)
```

## Step 6: Removing duplicates, if any.

Step to help remove any duplicate records from the dataset. 

```python
# Check for duplicates
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")

# Remove duplicate rows
df = df.drop_duplicates()

# Verify no duplicates remain
assert df.duplicated().sum() == 0
```

## Step 7: Standardizing the Date Column

This step is to convert the date column to a standardized format. This will be helpful when it comes to combining this data with our other sources.

#### Step 1: Connect to API. Store data into a dataframe. 

```python
import requests
import pandas as pd
import time
from scipy import stats
import numpy as np

# Your OpenWeatherMap API key (Ensure you have access to the One Call API)
api_key = '16684412aaa7fae5597f41342217afd6'  # Replace with your actual API key
lat, lon = 40.7128, -74.0060  # Latitude and longitude for New York City
start_date = '2015-01-01'  # Example start date
end_date = '2016-12-30'  # Example end date

# Function to convert date to Unix timestamp
def date_to_unix(date):
    return int(pd.Timestamp(date).timestamp())

# Generate a list of timestamps for each day in the range
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
timestamps = [date_to_unix(date) for date in date_range]

# Initialize an empty list to hold the weather data
weather_data = []

# Pull data from API
base_url = "https://api.openweathermap.org/data/3.0/onecall/timemachine"

for timestamp in timestamps:
    url = f"{base_url}?lat={lat}&lon={lon}&dt={timestamp}&appid={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        daily_data = data.get('data', [None])[0]  # Get the first element of 'data' list or None
        if daily_data is None:
            print(f"No data available for timestamp: {timestamp}")
        else:
            daily_data['timestamp'] = timestamp  # Add the timestamp to each record
            weather_data.append(daily_data)
    else:
        print(f"Failed to fetch data for timestamp: {timestamp}. Status code: {response.status_code}")
    # Adding sleep to comply with the API rate limit
    time.sleep(1)

# Convert the JSON data to a pandas DataFrame
df_weather = pd.json_normalize(weather_data)

df_weather.head()
```

#### Step #2: Replace Headers

```python
df_weather.rename(columns={
    'timestamp': 'Date',
    'temp': 'Temperature',
    'feels_like': 'Feels_Like',
    'pressure': 'Pressure',
    'humidity': 'Humidity',
    'dew_point': 'Dew_Point',
    'clouds': 'Cloud_Coverage',
    'visibility': 'Visibility',
    'wind_speed': 'Wind_Speed',
    'wind_deg': 'Wind_Direction',
    'weather': 'Weather_Description',
    'sunrise': 'Sunrise',
    'sunset': 'Sunset',
    'rain.1h': 'Rain_In',
    'wind_gust': 'Wind_Gust',
    'snow.1h': 'Snow_In',
    'dt': 'Unix_Timestamp'
}, inplace=True)

df_weather.head()
```

#### Step #3: Format Data into a More Readable Format

```python
df_weather['Date'] = pd.to_datetime(df_weather['Date'], unit='s')
df_weather['Sunset'] = pd.to_datetime(df_weather['Sunset'], unit='s')
df_weather['Sunrise'] = pd.to_datetime(df_weather['Sunrise'], unit='s')

df_weather.head()
```

#### Step #4: Add Temperature in Celsius and Fahrenheit

```python
df_weather['Temperature_C'] = df_weather['Temperature'] - 273.15
df_weather['Temperature_F'] = df_weather['Temperature_C'] * 9/5 + 32

df_weather.head(2)
```

#### Step #5: Identify Outliers

```python
# Using Z-score method to identify outliers in the Temperature column
z_scores = np.abs(stats.zscore(df_weather['Temperature']))
outliers = df_weather[z_scores > 3]
print("Outliers detected:\n", outliers)
```

#### Step #6: Fix Casing or Inconsistent Values

```python

df_weather['Weather_Description'] = df_weather['Weather_Description'].apply(lambda x: x[0]['description'].lower())
```

#### Step #5: Deal with Missing Data

```python
df_weather.ffill(inplace=True)

# Display the cleaned and formatted data
df_weather.head()
```
